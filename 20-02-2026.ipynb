## Lambda Functions 

```python
add = lambda a, b: a + b
print(add(2, 3))
```

Sorts list using second element.
Lambda used as key function.

```python
mul = lambda a,b,c: a*b*c
print(mul(2,3,4))
```

## Recursion 

```python
def fact(n):
    if n == 1:
        return 1
    return n * fact(n-1)
print(fact(5))
```

Computes factorial recursively.
Function calls itself with smaller input.

```python
def sum_n(n):
    if n == 0:
        return 0
    return n + sum_n(n-1)
print(sum_n(5))
```
