```python
person = {'name': 'Rimsha', 'Age': 21, 'phone_number': ['03285780400']}
print("Dictionary: ")
print(person) 

person["phone_number"].append('03019579061')

print("new dictionary:")
print(person)
```

```python
person = {'name': 'Rimsha', 'Age': 21, 'PH': ['03285780400']}
print("Dictionary:")
print(person)

person["PH"].append('03019579061')

print("New Dictionary:")
print(person)

person ['name']=('Noor')
print("New Dictionary 2: ")
print(person)

del person['Age']
print('new dictionary after age delete:')
print(person)
```
